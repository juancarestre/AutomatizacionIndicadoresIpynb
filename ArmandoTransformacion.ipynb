```python

class TransformacionRow(object):
    
    def __init__(self,app,sumaDefectos,sumaCasosPrueba,FileName):
        
    
        self.Anho=datetime.now().year
        self.SPRINT=getNextSprintNumber(transformacionDataSet)
        self.PITTs=FileName.split('/')[-1].split('.')[0]
        self.Aplicacion=app
        self.TxE2Eobjetivo=''
        self.TxE2EActivas=''
        self.CoberturaE2E=''
        self.TxRegobjetivo=''
        self.TxRegActivas=''
        self.CoberturaReg=''
        
        self.CPManuales=0
        self.CPAutomáticos=0
        self.CPTotales=0
        
        for appsEnListaCasosPrueba in sumaCasosPrueba.index.get_level_values(0).tolist():
            if appsEnListaCasosPrueba in app:
                duda=sumaCasosPrueba.groupby(['REL_NAME']).get_group(appsEnListaCasosPrueba)
                values=duda.values.tolist()[0]      
                self.CPManuales=values[0]
                self.CPAutomáticos=values[1]
                self.CPTotales=self.CPManuales+self.CPAutomáticos
                
        self.DefectosManuales=0
        self.DefectosAutomáticos=0
        self.DefectosTotales=0
        self.ImpactoAlto=0
        self.ImpactoMedio=0
        self.ImpactoBajo=0
        
        for appsEnListaDefectos in sumaDefectos.index.get_level_values(0).tolist():  
            if appsEnListaDefectos in app:
                duda=sumaDefectos.groupby(['Application']).get_group(appsEnListaDefectos)
                values=duda.values.tolist()[0]      
                self.DefectosManuales=values[0]
                self.DefectosAutomáticos=values[1]
                self.DefectosTotales=self.DefectosManuales+self.DefectosAutomáticos
                self.ImpactoAlto=values[2]
                self.ImpactoMedio=values[3]
                self.ImpactoBajo=values[4]
                
        self.SPRINT2='({})'.format(self.SPRINT)

    def getTransformacionRow(self):
        return [self.Anho,self.SPRINT,self.PITTs,self.Aplicacion,self.TxE2Eobjetivo,self.TxE2EActivas,self.CoberturaE2E,self.TxRegobjetivo,self.TxRegActivas,self.CoberturaReg,self.CPManuales,self.CPAutomáticos,self.CPTotales,self.DefectosManuales,self.DefectosAutomáticos,self.DefectosTotales,self.ImpactoAlto,self.ImpactoMedio,self.ImpactoBajo,self.SPRINT2]
    
    def getSoporteRow(self):
        return [self.Anho,self.SPRINT,self.PITTs,self.Aplicacion,self.TxE2Eobjetivo,self.TxE2EActivas,self.CoberturaE2E,self.TxRegobjetivo,self.TxRegActivas,self.CoberturaReg,self.CPManuales,self.CPAutomáticos,self.CPTotales,self.DefectosManuales,self.DefectosAutomáticos,self.DefectosTotales,self.ImpactoAlto,self.ImpactoMedio,self.ImpactoBajo,self.SPRINT2]
    
    def __str__(self):
        return str([self.Anho,self.SPRINT,self.PITTs,self.Aplicacion,self.TxE2Eobjetivo,self.TxE2EActivas,self.CoberturaE2E,self.TxRegobjetivo,self.TxRegActivas,self.CoberturaReg,self.CPManuales,self.CPAutomáticos,self.CPTotales,self.DefectosManuales,self.DefectosAutomáticos,self.DefectosTotales,self.ImpactoAlto,self.ImpactoMedio,self.ImpactoBajo,self.SPRINT2])
```

```python
resourcesPath="Resources/"

transformacionFile='baseTransformacion/TRANSFORMACION.xls'
soporteFile='baseSoporte/CERTIFICACIÓN.xls'



casosPruebaPath='Queries Qc/QC 12/'
defectosPath='Queries Qc/QC 12/DEFECTOS/'

defectosPath=os.path.join(resourcesPath,defectosPath)
casosPruebaPath=os.path.join(resourcesPath,casosPruebaPath)

defectosFiles=[]
casosPruebaFiles=[]

transformacionFilePath=os.path.join(resourcesPath,transformacionFile)
soporteFilePath=os.path.join(resourcesPath,soporteFile)


for file in os.listdir(defectosPath):
    if file.endswith(".xlsx"):
        defectosFiles.append(os.path.join(defectosPath, file))

for file in os.listdir(casosPruebaPath):
    if file.endswith(".xlsx"):
        casosPruebaFiles.append(os.path.join(casosPruebaPath, file))
```

```python
##Agregar a utils

def getNextSprintNumber(baseDataSet):
    return (max(baseDataSet['SPRINT'].tolist())+1)
```

```python
defectosDataFrames=[]
for file in defectosFiles:
    defectosDataFrame=pd.read_excel(file)
    defectosDataFrames.append(defectosDataFrame)

    
casosPruebasDataFrames=[]
for file in casosPruebaFiles:
    casosPruebaDataFrame=pd.read_excel(file)
    casosPruebasDataFrames.append(casosPruebaDataFrame)
```

```python
def cleanDefectosDataSet(defectosDataset):
    dummiesDefectosCanales=pd.get_dummies(columns=['Business Impact','Automatic Defect'],data=defectosCanalesDataSet)
    dummiesDefectosCanales=dummiesDefectosCanales.drop(columns=['Defect ID'])
    a=dummiesDefectosCanales.groupby(['Automatic Defect_N']).get_group(1)
    b=dummiesDefectosCanales.groupby(['Automatic Defect_N']).get_group(0)
    a['Business Impact_A-High']=0
    a['Business Impact_B-Medium']=0
    a['Business Impact_C-Low']=0
    defectosCanalesLimpios=a.append(b)
    return defectosCanalesLimpios
```

```python
def cleanCasosPruebaDataSet(casosPruebaDataSet,consolidado):
    
    releasesDataSet=casosPruebaDataSet['REL_NAME'].tolist()
    nombresDeAplicativo=casosPruebaDataSet['APLICATIVO'].tolist()

    casosPruebaCanalesLimpioDataSet = casosPruebaDataSet

    casoPruebaCanalesDataSet=[]
    for idx, val in enumerate(releasesDataSet):
        if consolidado=='Transformacion': casoPruebaCanalesDataSet.append('{} {}'.format(val.split('_')[0],nombresDeAplicativo[idx]))
        elif consolidado=='Soporte': casoPruebaCanalesDataSet.append('{}'.format(nombresDeAplicativo[idx]))
            
    casosPruebaCanalesLimpioDataSet=casosPruebaCanalesLimpioDataSet.reset_index().drop(columns=['index'])
    casosPruebaCanalesLimpioDataSet['REL_NAME']=pd.DataFrame(casoPruebaCanalesDataSet)
    casosPruebaCanalesLimpioDataSet=casosPruebaCanalesLimpioDataSet.drop(columns=['CY_CYCLE_ID'])
    
    return casosPruebaCanalesLimpioDataSet
```

```python
def getAppsToRows(listaCasosPruebaSumadosDataset,listaDefectosSumadosDataset):
    
    listaAppsCasosPrueba=listaCasosPruebaSumadosDataset.index.tolist()
    listaAppsDefectos=listaDefectosSumadosDataset.index.get_level_values(0).tolist()
    toAggregate=listaAppsDefectos
    total=listaAppsCasosPrueba
    for appCasoPrueba in listaAppsCasosPrueba:
        for appDefectos in listaAppsDefectos:
            if appDefectos in appCasoPrueba:
                toAggregate.remove(appDefectos)

    total=total+toAggregate
    return total
```

```python
def startProcess(sprintNumber,excelFileName,consolidado):
    
    if consolidado=='Transformacion': DataSet=pd.read_excel(transformacionFilePath)
    elif consolidado=='Soporte': DataSet=pd.read_excel(soporteFilePath)
        
    transformacionDataSetNuevo=DataSet.drop(DataSet.index, inplace=False)
    bigdata = DataSet.append(transformacionDataSetNuevo, ignore_index=True)

    for file in range(len(defectosDataFrames)):

        defectosDataset=defectosDataFrames[file]
        casosPruebasDataset=casosPruebasDataFrames[file]

        try:
            defectosDataset=defectosDataset.groupby(['Detected By']).get_group(consolidado)
            casosPruebasDataset=casosPruebasDataset.groupby(['Detected By']).get_group(consolidado)
            defectosDataset=defectosDataset.groupby(['Detected in Cycle']).get_group('Sprint'+str(sprintNumber))
            casosPruebasDataset=casosPruebasDataset.groupby(['SPRINT']).get_group('Sprint'+str(sprintNumber))

            defectosDataset=cleanDefectosDataSet(defectosDataset)
            defectosDataset=defectosDataset.groupby(['Application','Detected in Cycle','Detected By']).sum()

            casosPruebasDataset=cleanCasosPruebaDataSet(casosPruebasDataset,consolidado)
            casosPruebasDataset=casosPruebasDataset.groupby(['REL_NAME']).sum()

            total=getAppsToRows(casosPruebasDataset,defectosDataset)

            rowToAggregate=[]
            for app in total:
                rowToAggregate.append(TransformacionRow(app,defectosDataset,casosPruebasDataset,defectosFiles[file]))

            transformacionDataSetNuevo=DataSet.drop(DataSet.index, inplace=False)

            for i in range(len(rowToAggregate)):
                if consolidado=='Transformacion':
                    transformacionDataSetNuevo.loc[i] = rowToAggregate[i].getTransformacionRow()
                elif consolidado=='Soporte':
                    transformacionDataSetNuevo.loc[i] = rowToAggregate[i].getSoporteRow()
                    
            bigdata = bigdata.append(transformacionDataSetNuevo, ignore_index=True)
            bigdata=bigdata.sort_values(by=['SPRINT'],ascending=False)
            bigdata=bigdata.reset_index()
            bigdata=bigdata.drop(columns=['index'])


        except:
            defectosDataset=defectosDataset.groupby(['Detected By'])
            casosPruebasDataset=casosPruebasDataset.groupby(['Detected By'])

    bigdata.to_excel(excelFileName,index=False)        

```

```python
startProcess(41,'consolidadoTransformacion.xlsx','Transformacion')
startProcess(41,'consolidadoSoporte.xlsx','Soporte')
```
